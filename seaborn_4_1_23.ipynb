```python
import seaborn as sns
df=sns.load_dataset('tips')
import warnings
warnings.filterwarnings('ignore')
df
```

```python
df.corr()         #correlation
```

```python
sns.heatmap(df.corr())          #express outcome (correlation) in pictorial form
                                #light pink:  better corelation
                                #pink
                                #purple
                                #black:       less correlation
```

```python
#jointplot
sns.jointplot(x='size',y='total_bill',data=df)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df)
```

```python
sns.jointplot(x='total_bill',y='day',data=df)
```

```python
#paiplot
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='day')
```

```python
#distplot
sns.distplot(df['tip'])         #the line in map :  range of distribution
```

```python
#distplot
sns.distplot(df['tip'],kde=False,bins=3)
```

```python
sns.distplot(df['tip'],kde=False,bins=5)
```

```python
# CATEGORIAL  PLOTS
#     countplot
#     barplot
#     boxplot
#     violinplot
```

```python
#COUNTPLOT
sns.countplot('sex',data=df)
```

```python
sns.countplot(df['sex'])
```

```python
sns.countplot('sex',hue='smoker',data=df)
```

```python
sns.countplot('sex',hue='day',data=df)
```

```python
sns.countplot('sex',hue='time',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
#BARPLOT
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
#BOXPLOT
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df)
```

```python
#VIOLIN PLOT
sns.violinplot('total_bill','day',data=df)
```

```python
sns.violinplot('total_bill','sex',data=df)
```

```python
sns.violinplot('total_bill','time',data=df)
```

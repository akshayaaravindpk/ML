```python
import matplotlib.pyplot as plt
x=[5,3,2,8]
y=[1,4,2,7]
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph Title')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[5,3,2,8]
y=[1,4,2,7]
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph Title')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[5,3,2,8]
y=[1,4,2,7]
plt.plot(x,y,'+')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph Title')
plt.show()
```

```python
import numpy as np
y=np.array([1,4,2,5,6])
plt.plot(y)
plt.show()
```

```python
import numpy as np
y=np.array([1,4,2,5,6])
plt.plot(y,'+:r')
#+- marker
#: ->line
#r ->color
plt.show()
```
